```python
import pandas as pd   
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import joblib
```

```python
# 2. Load and Explore Data
df = pd.read_csv('carprices.csv')
print("Dataset Preview:")
df.head()
```

## Data Visualization

```python
# Plot Mileage vs Sell Price
plt.figure(figsize=(10, 5))
plt.scatter(df['Mileage'], df['Sell Price($)'], alpha=0.6)
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.title('Mileage vs Sell Price')
plt.show()
```

```python
# Plot Age vs Sell Price
plt.figure(figsize=(10, 5))
plt.scatter(df['Age(yrs)'], df['Sell Price($)'], alpha=0.6, color='orange')
plt.xlabel('Age (years)')
plt.ylabel('Sell Price($)')
plt.title('Age vs Sell Price')
plt.show()
```

## 3. Prepare Data for Model

```python
# Prepare Features (X) and Target (y)
x = df[['Mileage', 'Age(yrs)']]
y = df['Sell Price($)']

print(f"Features shape: {x.shape}")
print(f"Target shape: {y.shape}")
```

## 4. Train Model

```python
# Split data into training and testing sets (60% train, 40% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)

print(f"Training set size: {x_train.shape[0]}")
print(f"Testing set size: {x_test.shape[0]}")
```

```python
# Train Linear Regression Model
model = LinearRegression()
model.fit(x_train, y_train)

print("Model trained successfully!")
print(f"Coefficients: Mileage={model.coef_[0]:.4f}, Age={model.coef_[1]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
```

## 5. Evaluate Model

```python
# Evaluate Model Performance
r2_score = model.score(x_test, y_test)
print(f"Model R Score on Test Set: {r2_score:.4f}")
```

```python
# Make Predictions
y_pred = model.predict(x_test)

# Compare predictions with actual values
comparison = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred,
    'Difference': abs(y_test.values - y_pred)
})

print(comparison.head(10))
```

## 6. Save and Load Model

```python
# Save the trained model
joblib.dump(model, 'car_price_model.pkl')
print("Model saved as 'car_price_model.pkl'")
```

```python
# Load the saved model
loaded_model = joblib.load('car_price_model.pkl')
print("Model loaded successfully!")
```

```python
# Test the loaded model
test_predictions = loaded_model.predict(x_test)
print("Predictions from loaded model:")
print(test_predictions[:5])
```
